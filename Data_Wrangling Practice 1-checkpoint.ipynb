```python
#699 Rows, but here we have 637 ID'S, so we have duplicates, that means many patients were duplicated or appeared more than one time
#we can isolate these patients and visualize their data

df[df.duplicated(subset = 'patient_id',keep=False)].sort_values('patient_id')
```

```python
#what if the inmsurance company asked clearly about how many visits were done for every duplicated patient 
#so it can price their rennvels accordingly??

repeat_patients = df.groupby(by = 'patient_id').size().sort_values(ascending = False)

```

```python
# it worth to pinpoint that patient 1182404 has 7 visitations (6 repetitions)
# now, assume that i need to remove patients that show up more than 2 times in the data set.
filtered_patients = repeat_patients[repeat_patients > 2].to_frame().reset_index()
filtered_df = df[~df.patient_id.isin(filtered_patients.patient_id)]
```

```python
#here i used the tilde(~). if i didn't include it we would get all patients that repeat more than twice. The tilde reverse the boeelean series in pandas

#and then we will have the resulting data frame of those who do NOTTTT repeat more than twice.
```

```python
# Reshaping data, here for example, we have 2 types in the same Dataset. (Doctor's name is categorical)
# this issue is very noisy for ML algorithms, exactly in pipelining data.
# we will need to call sci-kit learn here
# this is the simple solution
```

```python
# I discovered another way
# pivoting and merging technic
# let us start by creating a new dataframe WITH the categorical data.
categorical_df = df[['patient_id','doctor_name']]
categorical_df['doctor_count'] = 1
```

```python
# now we add an extra column to identify which doctor a patient deals with.
doctors_one_hot_encoded = pd.pivot_table(categorical_df,
                                        index = categorical_df.index,
                                        columns = ['doctor_name'],
                                        values = ['doctor_count']
                                        )
doctors_one_hot_encoded = doctors_one_hot_encoded.fillna(0)
```

```python
doctors_one_hot_encoded
```

```python
doctors_one_hot_encoded.columns = doctors_one_hot_encoded.columns.droplevel()
```
