```python
# Write a program that displays the first 100 palindormic prime numbers.

# each 10 numbers per one line



#fct to check if it is prime



def isPrime(number):

  divisor = 2

  while divisor <= number / 2:

    if number % divisor == 0:

      #if true it means that the number isn't prime

      return False

    divisor += 1

  return True #means that the number is prime



#now we will define a ffct to reverse an int 123 --> 321



def reverse(number):

  result = 0

  while number != 0:

    remainder = number % 10

    result = result * 10 + remainder

    number = number // 10

   

  return result



#now we will define a fct to check a palindrome

def isPalindrome(number):

  return number == reverse(number)



counter = 1

x = 2

#loop for the firs 100 Palindromic prime numbers



while counter < 101:

  if isPrime(x) and isPalindrome(x):

    if counter % 15 != 0:

      print(x, end= ' ')

    else:

      print(x)

    counter += 1

  x += 1
```
