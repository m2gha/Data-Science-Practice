```python
#Regular Expression (super powerfull in complex filtering) Regex engine
import re
df.loc[df['Type 1'].str.contains('Fire'||'Grass',regex=True) #or not working with str.contains
```

```python
#overlook or drop caps
df.loc[df['Type 1'].str.contains('fire || grass',flags=re.I,regex=True)]#or not working with str.contains
```

```python
#filter out all names that start with pi followed or not by a on or more symbol
df.loc[df['Name'].str.contains('^pi[a-z]*',flags = re.I, regex=True)]
# extremely important
```

```python
df.loc[df["Type 2"].str.contains("^F[a-z]*",na=False,flags=re.I,regex=True)]
```

```python
# repeat regex
import re
df.loc[df['Type 1'].str.contains('Fire', regex=True)] #worked
```

```python
df.loc[df['Type 1'].str.contains('Fire',regex = True) | df['Type 1'].str.contains('Grass',regex = True)]#or in here worked perfectly
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass',flags=re.I,regex=True)]#or in here too worked
```
