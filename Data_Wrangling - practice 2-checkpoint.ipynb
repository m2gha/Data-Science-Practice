```python
import pandas as pd

pd.options.display.max_columns = None

pd.options.display.max_rows = 40



filename= 'visua_data_canc.csv'

df = pd.read_csv(filename)
```

```python
# I discovered another way
# pivoting and merging technic
# let us start by creating a new dataframe WITH the categorical data.
categorical_df = df[['patient_id','doctor_name']]
categorical_df.loc[:,'doctor_count'] = 1
categorical_df
```

```python
# now we add an extra column to identify which doctor a patient deals with.
doctors_one_hot_encoded = pd.pivot_table(categorical_df,
                                        index = categorical_df.index, #merging it
                                        columns = ['doctor_name'],
                                        values = ['doctor_count']
                                        )
doctors_one_hot_encoded = doctors_one_hot_encoded.fillna(0)

doctors_one_hot_encoded
```

```python
#here the pivot table is generated correctly but it isn't beneficial unless we merge it with the original table
#to do this: some DS use leftmerge. I personally not use it, so i cobine directly.

combined_df = pd.merge(df, doctors_one_hot_encoded, left_index = True, right_index = True, how = 'left')
combined_df
```

```python
#the column doctor's name is irrelevantt after the merge so we have to drop it.
combined_df = combined_df.drop(columns = ['doctor_name'])
combined_df
```

```python
# one of the most important features of data wrangling, is row-sise and columns-wise operations
# here we are asked to create a new columns that categorises (classifies) a patient's cell as normal or abnormal.

def celltypelabel(x):
    if((x['cell_size_uniformity']>5) & (x['cell_shape_uniformity']>5)):
        return('normal')
    else:
        return('abnormal')
```

```python
combined_df['cell_typelabel'] = combined_df.apply(lambda x: celltypelabel(x), axis =1)
combined_df
```
